```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the salary_data dataset
Salary_Data_df = pd.read_csv('Salary_Data.csv')

# Display the first few rows of the dataset
print(Salary_Data_df.head())

# Get a summary of the dataset
print(Salary_Data_df.info())

# Get basic statistics of numerical columns
print(Salary_Data_df.describe())
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(Salary_Data_df['YearsExperience'])
plt.title('YearsExperience Distribution')

plt.subplot(1, 3, 2)
sns.histplot(Salary_Data_df['Salary'])
plt.title('Salary Distribution')


plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = Salary_Data_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot countplots for gender and survival
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='YearsExperience', data=Salary_Data_df)
plt.title('Experience')

plt.subplot(1, 2, 2)
sns.countplot(y='Salary', data=Salary_Data_df)
plt.title('Salary')

plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(Salary_Data_df['YearsExperience'], Salary_Data_df['Salary'], color='blue', alpha=0.5)
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
